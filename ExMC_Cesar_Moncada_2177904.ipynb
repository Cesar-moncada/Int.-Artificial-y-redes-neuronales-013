### Parte 1: Carga y Limpieza (Pandas y NumPy)

* Carga de dataset, importacion de librerias y primeras 5 filas

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

# Mostrar primeras 5 filas
df.head()
```

* Tipo de datos

```python
# Tipos de datos
df.dtypes
```

* Valores nulos en horsepower

```python
# Verificar valores nulos en horsepower
df['horsepower'].isnull().sum()
```

* Reemplezar nulos con mediana

```python
# Reemplazar nulos con la mediana
mediana_hp = np.nanmedian(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mediana_hp)

# Verificación
df['horsepower'].isnull().sum()
```

* Nueva columna peso/potencia

```python
# Crear nueva columna relación peso/potencia
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']

df.head()
```

### Parte 2: Visualización de Datos (Matplotlib)

* Histograma



```python
plt.figure()
plt.hist(df['acceleration'])
plt.title('Distribución de la aceleración')
plt.xlabel('Aceleración')
plt.ylabel('Frecuencia')
plt.show()
```

* Scatter Plot

```python
plt.figure()
plt.scatter(df['displacement'], df['horsepower'])
plt.title('Desplazamiento vs Caballos de Fuerza')
plt.xlabel('Displacement')
plt.ylabel('Horsepower')
plt.show()
```

* Gráfico de Barras

```python
# Promedio de mpg por origen
mpg_promedio = df.groupby('origin')['mpg'].mean()

plt.figure()
mpg_promedio.plot(kind='bar')
plt.title('Promedio de MPG por Origen')
plt.xlabel('Origen')
plt.ylabel('MPG Promedio')
plt.show()
```
